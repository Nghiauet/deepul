```python
from diffusers import AutoPipelineForText2Image
import torch

pipe_txt2img = AutoPipelineForText2Image.from_pretrained(
    "dreamlike-art/dreamlike-photoreal-2.0", torch_dtype=torch.float16, use_safetensors=True
).to("cuda")

prompt = "Hyper-realistic concept car design, sleek aerodynamic sports coupe with dramatic swooping lines, carbon fiber body, gull-wing doors partially open, set against an industrial architectural background at golden hour, professional automotive photography, studio lighting highlighting the metallic paint finish, ultra-detailed engineering, precision modeling, ray-traced reflections, unreal engine 5 render, 8K resolution, automotive design showcase, concept by top industrial designer"
generator = torch.Generator(device="cpu")
image = pipe_txt2img(prompt, generator=generator).images[0]
image
```
